```python
import robin_stocks as rh
import cfg
import pandas as pd
from yahoofinancials import YahooFinancials

pd.set_option('display.max_columns', None)
```

```python
rh.login(cfg.email, cfg.password)
```

```python
stocks = [rh.stocks.get_instrument_by_url(stock["instrument"])["symbol"] for stock in rh.account.get_watchlist_by_name()]
```

```python
stockData = {}
yahoo_financials = YahooFinancials(stocks)
data = yahoo_financials.get_historical_price_data(start_date="2018-01-01", end_date="2020-05-08", time_interval="daily")

metrics = ["close", "volume", "high", "low"]
for stock in stocks:
    print(stock)
    stockData[stock] = pd.DataFrame({stock + "_"+i: {x['formatted_date']: x[i] for x in data[stock]['prices']} for i in metrics})
```

```python
data = stockData[list(stockData.keys())[0]]
for ticker in list(stockData.keys())[1:]:
    data = data.merge(stockData[ticker], left_index=True, right_index=True, how="outer")

data.head()
```

```python
for stock in stocks:
    data["5DayMA_"+stock] = data[stock+"_close"].rolling(window=5).mean()
    data["10DayMA_"+stock] = data[stock+"_close"].rolling(window=10).mean()
    data["20DayMA_"+stock] = data[stock+"_close"].rolling(window=20).mean()
    data["50DayMA_"+stock] = data[stock+"_close"].rolling(window=50).mean()
    data["100DayMA_"+stock] = data[stock+"_close"].rolling(window=100).mean()
    data["200DayMA_"+stock] = data[stock+"_close"].rolling(window=200).mean()

data.dropna(inplace=True)
```

```python
class Stock:
    def __init__(self, ticker, numberOfShares, price):
        self.ticker = ticker
        self.numberOfShares = numberOfShares
        self.price = price
    
    def __repr__(self):
        return(self.ticker + ": " + str(self.numberOfShares) + " @ $"+str(self.price))

cash = 0 
portfolio = []
```

```python
data = data.reset_index()
data.head()
```

```python
def getAssetValue(cash, portfolio, index):
    assets = cash
    for stock in portfolio:
        assets += stock.numberOfShares * data.loc[index, stock.ticker+"_close"]
    return(assets)
```

```python
import random as rd
cash = 1000
portfolio = []
spyportfolio = []
data["portfolioValue"] = 0

for i, row in data.tail(len(data)-1).iterrows():
    rd.shuffle(stocks)
    for stock in stocks:
        shortTerm = "5DayMA_"+stock
        longTerm = "10DayMA_" + stock
        close = stock + "_close"
        if row[shortTerm] >= row[longTerm] and data.loc[i-1, shortTerm] < data.loc[i-1, longTerm] and cash > data.loc[i, close]:
            purchasePrice = data.loc[i, shortTerm]
            numberOfShares = int(cash/purchasePrice)
            purchased = Stock(stock, numberOfShares, purchasePrice)
            portfolio.append(purchased)
            cash -= purchasePrice * numberOfShares
            #print("Purchasing: "+ str(purchased))
        elif row[shortTerm] <= row[longTerm] and data.loc[i-1, shortTerm] > data.loc[i-1, longTerm]:
            sold = []
            for position in portfolio:
                if position.ticker == stock:
                    sold.append(position)
                    cash += data.loc[i, longTerm] * position.numberOfShares
                    print("Selling: " + str(position) + " for $" + str(data.loc[i, longTerm]))
            for item in sold:
                portfolio.remove(item)
    data.loc[i, "portfolioValue"] = getAssetValue(cash, portfolio, i)
```
